## 1.统计翻译简介

比如将“今晚的课程很有意思”翻译成英文：<br>
1.分词： 将中文“今晚的课程很有意思”分词得到：今晚、的、课程、有意思<br>
2.查找字典：对应成tonight、of、course、interesting(关键步骤)<br>
3.组合：tonight of interesting course\tonight interesting of course\\.....（造成计算复杂度增大的步骤）<br>
4.计算概率：根据LM（language Model）计算句子概率（关键步骤）<br>
5.选取概率最大的那个句子（course of tonight is interesting）


对于第三步，可以用维特比算法解决。
