## 1.归并排序

归并排序算法采用的是分治策略。这里我们将用python实现归并排序然后介绍如何计算归并排序的时间复杂度。

```python
#merge（合并）过程;注意len(arr)奇偶性的不同程序可能要改动，这里我们以奇数为例。
def merge_arr():
    global arr
    mid = int(len(arr)/2)
    temp = [0]*len(arr)
    
    i = 0
    k = 0
    j = mid+1 
    
    while i<mid+1 and j<len(arr):
        #这里用“<=”而不用“<” 是为了保证排序算法的稳定
        if arr[i]<=arr[j]:
            temp[k]=arr[i]
            i+=1
            k+=1
        else:
            temp[k] = arr[j]
            j+=1
            k+=1
    temp[k:] = arr[i:mid+1] if i<mid+1 else arr[j:]
    arr[:] = temp[:]
    return arr
```

```python
arr=[1,4,7,8,3,6,9]
print(merge_arr())
print(arr)
```

在归并排序的过程中我们经常要排序数组的一部分而不是整个数组，所以我们需要对上述算法进行改进。

```python
def merge_arr_pro(left_point, right_point, right_bound):
    global arr
    mid = right_point - 1
    temp = [0]*(right_bound-left_point+1)
    
    i = left_point
    k = 0
    j = right_point
    
    while i<mid+1 and j<right_bound+1:
        #这里用“<=”而不用“<” 是为了保证排序算法的稳定
        if arr[i]<=arr[j]:
            temp[k]=arr[i]
            i+=1
            k+=1
        else:
            temp[k] = arr[j]
            j+=1
            k+=1
    temp[k:] = arr[i:mid+1] if i<mid+1 else arr[j:right_bound+1]
    arr[left_point:right_bound+1] = temp[:]
    return arr
```

```python
arr=[1,4,7,8,3,6,9]
print(merge_arr_pro(1,4,5))
```

```python
def sort_arr(left, right):
    # 若数组只有一个元素
    if left == right:
        return
    
    mid = int((left+right)/2)
    #对左数组排序
    sort_arr(left, mid)
    #对右数组排序
    sort_arr(mid+1, right)
    merge_arr_pro(left, mid+1, right)
```

```python
arr = [1,4,7,8,3,6,9]
sort_arr(0,6)
print(arr)
```
